### Problema 10

The sum of the primes below 10 is 2 + 3 + 5 + 7 = 17.

Find the sum of all the primes below two million.

```python
limite = 2000000
criba = [True] * limite

# Como cero y uno no son primos, los marcamos como falsos
criba[0] = criba[1] = False

# Sólo se necesita comprobar hasta la raíz cuadrada del límite
raiz = int(limite**0.5)

for i in range(2, raiz + 1):
    if criba[i]:
        for j in range(i * i, limite, i):
            criba[j] = False

# Se suman todos los true en la criba, que representan los números primos
suma_primos = sum(i for i, es_primo in enumerate(criba) if es_primo)

print(suma_primos)
```
