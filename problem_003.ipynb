### Problem 3

The prime factors of 13195 are 5,7,13 and 29.

What is the largest prime factor of the number 600851475143?

```python
limit = 600851475143

def largest_prime_factor(n):
    largest = -1

    # Se comprueba el factor 2
    while n % 2 == 0:
        largest = 2
        n //= 2

    # Se comprueba los factores impares
    for i in range(3, int(n**0.5) + 1, 2):
        while n % i == 0:
            largest = i
            n //= i

    # Si n es un número primo mayor que 2
    if n > 2:
        largest = n

    return largest

largest_prime_factor(limit)
```

Al dividir iterativamente $n$ por sus factores más pequeños $i$, se garantiza que cualquier factor futuro encontrado no pueda ser compuesto (ya que sus divisores primos habrían sido eliminados previamente). Por lo tanto, no es necesario un test de primalidad explícito.
