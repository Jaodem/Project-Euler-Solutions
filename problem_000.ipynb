### Problem Zero

A number is a perfect square, or a square number, if it is the square of a positive integer.
For example, 25 is a square number because 5² =5 ×5 =25; it is also an odd square.

The first 5 square numbers are: 1,4,9,16,25, and the sum of the odd squares is 1 +9 +25 =35.

Among the first 312 thousand square numbers, what is the sum of all the odd squares?

```python
# Se quiere obtener la suma de i²
# Donde 'i va desde 1 hasta 321.000
# Pero solamente se sumarán los impares (step de 2)

limite = 312000
suma = sum(i**2 for i in range(1, limite + 1, 2))

print(f'La suma de los cuadrados impares hasta {limite} es: {suma}')
```
